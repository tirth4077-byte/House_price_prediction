```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('../data/train_cleaned.csv')
print("Shape:", df.shape)
```

```python
X = df.drop('SalePrice', axis=1)
y = df['SalePrice']

print("Features shape:", X.shape)
print("Target shape:", y.shape)
```

```python
# Total area of house
X['TotalSF'] = X['TotalBsmtSF'] + X['1stFlrSF'] + X['2ndFlrSF']

# Age of house when sold
X['HouseAge'] = X['YrSold'] - X['YearBuilt']

# Did the house get remodelled?
X['IsRemodeled'] = (X['YearRemodAdd'] != X['YearBuilt']).astype(int)

# Total bathrooms
X['TotalBath'] = X['FullBath'] + (X['HalfBath'] * 0.5)

print("Shape after new features:", X.shape)
```

```python
cat_cols = X.select_dtypes(include=['object']).columns
print("Categorical columns:", len(cat_cols))

le = LabelEncoder()
for col in cat_cols:
    X[col] = le.fit_transform(X[col].astype(str))

print("Done encoding!")
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print("Shape after scaling:", X_scaled.shape)
```

```python
X_scaled['SalePrice'] = y.values
X_scaled.to_csv('../data/train_final.csv', index=False)
print("Final data saved!")
```

```python
X['SalePrice'] = y.values
X.to_csv('../data/train_features_unscaled.csv', index=False)
print("Unscaled features saved!")
```
